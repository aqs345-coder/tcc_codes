# Aplicações de Filtro de Kalman para Ciência de Dados

Este repositório contém implementações passo a passo do **Filtro de Kalman** aplicado a problemas de rastreamento e estimação de estado, com base no tutorial [Kalman Filter Net](https://kalmanfilter.net/PT/default_pt.aspx).

## Estrutura

- **`index.ipynb`** — este arquivo, visão geral do repositório.
- **`exemplos/`** — implementações iniciais dos exemplos do tutorial.
- **`tutoriais/`** — notebooks com implementações detalhadas dos exemplos.

### Exemplos disponíveis

| # | Exemplo | Descrição |
|---|---------|-----------|
| 1 | [Radar unidimensional (Kalman matricial)](exemplos/examplo_1.ipynb) | Rastreamento de aeronave (alcance e velocidade) com Filtro de Kalman em forma matricial |

### Tutoriais disponíveis

#### 1 — Introdução

| # | Tutorial | Descrição |
|---|----------|-----------|
| 1 | [Filtro Alpha — Peso de uma barra de ouro](tutoriais/1_introducao/barra_de_ouro_otimizado.ipynb) | Filtro alpha com ganho decrescente ($\alpha_n = 1/n$) para estimar o peso real de uma barra de ouro a partir de 10 medições ruidosas |
| 2 | [Filtro Alfa-Beta — Rastreamento de aeronave com velocidade constante](tutoriais/1_introducao/aeronave_vconst.ipynb) | Filtro alfa-beta para estimar posição e velocidade de uma aeronave a partir de medições ruidosas de radar |
| 3 | [Filtro de Kalman — Estimando a altura de um prédio](tutoriais/1_introducao/altura_predio.ipynb) | Filtro de Kalman escalar sem modelo dinâmico ($Q = 0$) para estimar a altura real de um prédio |
| 4 | [Filtro de Kalman — Temperatura de um líquido](tutoriais/1_introducao/temp_liquido.ipynb) | Filtro de Kalman escalar com ruído de processo ($Q > 0$) para estimar a temperatura aproximadamente constante de um líquido |
| 5 | [Filtro de Kalman — Temperatura de um líquido aquecido (subestimado)](tutoriais/1_introducao/temp_liquido_aquecido.ipynb) | Mesmo filtro aplicado a um líquido em aquecimento; **falha** porque o ruído de processo ($Q = 0{,}0001$) é muito pequeno para acompanhar a variação |
| 6 | [Filtro de Kalman — Temperatura de um líquido aquecido (corrigido)](tutoriais/1_introducao/temp_liquido_aquecido_corrigido.ipynb) | Versão **corrigida** com $Q = 0{,}15$, permitindo rastrear corretamente a temperatura crescente |
